# P2.12
sol)  
$$
Y(s)=K\frac{1}{s+50}R(s)
$$  
$r(t)$가 단위 계단 함수이기 때문에 $r(t)=1(t\geq0)$이고, $R(s)=\frac{1}{s}$이다.  
따라서 $Y(s)$는 다음과 같다.  
$$
Y(s)=K\frac{1}{s+50}\frac{1}{s}\\
sY(s)=K\frac{1}{s+50}
$$

$\lim_{t \rightarrow \infty}y(t)=\lim_{t \rightarrow \infty}sY(s)=1$로 수렴해야 하므로  
$\lim_{t \rightarrow \infty}sY(s)=\lim_{t \rightarrow \infty}K\frac{1}{s+50}=\frac{K}{50}=1$

따라서 결론은 다음과 같다.  
$$
K=50
$$

#P.26
sol)  
1) 질량이 M인 물체
$$
M\ddot{x}(t)+b\{\dot{x}(t)-\dot{y}(t)\}+k\{x(t)-jy(t)\}dt=F(t)\\
s^2MX(s)+bs\{X(s)-Y(s)\}+k\{X(s)-Y(s)\}=F(s)\\
(s^2M+bs+k)X(s)-(bs+k)Y(s)=F(s)\\
X(s)=\frac{(bs+k)Y(s)+F(s)}{s^2M+bs+k}
$$  

2) 질량이 m인 물체  
$$
m\ddot{y}(t)=-b\{\dot{y}(t)-\dot{x}(t)\}-k\{y(t)-x(t)\}\\
s^2mY(s)=-bs\{Y(s)-X(s)\}-k\{Y(s)-X(s)\}\\
=(bs+k)X(s)-(bs+k)Y(s)\\
(bs+k)X(s)=(s^2m+bs+k)Y(s)\\
X(s)=\frac{s^2m+bs+k}{bs+k}Y(s)
$$  

1), 2)를 연립하면  
$$
\frac{(bs+k)Y(s)+F(s)}{s^2M+bs+k}=\frac{s^2m+bs+k}{bs+k}Y(s)\\
\frac{bs+k}{s^2M+bs+k}Y(s)-\frac{s^2m+bs+k}{bs+k}Y(s)=-\frac{1}{s^2M+bs+k}F(s)
$$  
여기서 $bs+k=\alpha$로 치환하고 정리하면  
$$
\frac{\alpha}{s^2M+\alpha}Y(s)-\frac{s^2m+\alpha}{\alpha}Y(s)=-\frac{1}{s^2M+\alpha}F(s)\\
\left(\frac{\alpha}{s^2M+\alpha}-\frac{s^2m+\alpha}{\alpha}\right)Y(s)=-\frac{1}{s^2M+\alpha}F(s)\\
\frac{\alpha^2-(s^4Mm+s^2M\alpha+s^2m\alpha+\alpha^2)}{s^2M\alpha+\alpha^2}Y(s)=-\frac{1}{s^2M+\alpha}F(s)\\
$$  
$$
\frac{Y(s)}{F(s)}=\frac{-\alpha(s^2M+\alpha)}{-(s^4Mm+s^2M\alpha+s^2m\alpha)(s^2M+\alpha)}\\
=\frac{\alpha}{s^4Mm+s^2M\alpha+s^2m\alpha}\\
=\frac{\frac{1}{mM}(bs+k)}{s^2\{s^2+\frac{1}{m}(bs+k)+\frac{1}{M}(bs+k)\}}
$$  
따라서 결론은 다음과 같다.  
$$
\frac{Y(s)}{F(s)}=\frac{\frac{1}{mM}(bs+k)}{s^2\{s^2+\frac{1}{m}(bs+k)+\frac{1}{M}(bs+k)\}}
$$

#P2.37
sol)  
(a)  
$$
K_1x(t)+m_1\ddot{x}(t)=K_2\{y(t)-x(t)\}\\
u(t)=K_2\{y(t)-x(t)\}+m_2\ddot{y}(t)
$$  
2차 미분식으로 정리하면  
$$
m_1\ddot{x}(t)=-(K_1+K_2)x(t)+K_2y(t)\\
m_2\ddot{y}(t)=-K_2\{y(t)-x(t)\}+u(t)
$$  

(b)  
$$
m_1=m_2=1, K_1=K_2=1
$$  
$$
\ddot{x}(t)=-2x(t)+y(t)\\
\ddot{y}(t)=x(t)-y(t)+u(t)
$$  
$$
s^2X(s)=-2X(s)+Y(s)\\
s^2Y(s)=X(s)-Y(s)+U(s)
$$  
$$
X(s)=\frac{1}{s^2+2}Y(s)\\
X(s)=(s^2+1)Y(s)-U(s)
$$  
두 식을 연립하면  
$$
\frac{1}{s^2+2}Y(s)=(s^2+1)Y(s)-U(s)\\
U(s)=\left(s^2+1-\frac{1}{s^2+2}\right)Y(s)\\
=\frac{(s^2+1)(s^2+2)-1}{s^2+2}Y(s)\\
=\frac{s^4+3s^2+1}{s^2+2}Y(s)
$$  
따라서 결론은 다음과 같다.  
$$
\frac{Y(s)}{U(s)}=\frac{s^2+2}{s^4+3s^2+1}
$$
